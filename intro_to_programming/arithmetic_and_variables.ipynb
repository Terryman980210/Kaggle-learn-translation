### Creating variables

The next code cell creats a variable named test_var and assigns it the value that we get when we add 5 to 4.
We then print the value that is assigned to the variable, which is 9.

訳）次のコードセルはtest_varという名前の変数を作り、5に4を足して得られる値をその変数に割り与てます。
そのあと、変数に割り当てられた値である9を出力する。

```python
# Create a variable called test_var and give it a value of 4 + 5
test_var = 4 + 5

# Print the value of test_var
print(test_var)
```

In general, to work with a variable, you need to begin by selecting the name you want to use.
Variable names are ideally short and descriptive.
They also need to satisfy several requirements:

- They can't have spaces (e.g., test var is not allowed)
- They can only include letter, numbers, and underscores (e.g., test_var! is not allowed)
- They have to start with a letter or underscore (e.g., 1_var is not allowed)

Then, to create the variable, you need to use = to assign the value that you want it to have.

You can always take a look at the value assigned to the variable by using print() and putting the name of the variable in parentheses.

Overtime, you'll learn how to select good names for Python variables. It's completely fine for it to feel uncomfortable now, and the best way to learn is just by viweing a lot of Python code!

訳）一般的に、変数を扱うときは、まず使いたい名前を決める必要があります。
変数名は短くて内容の表しているものが理想です。
また、変数名には次の条件があります。
 - スペースを入れてはいけない（例：test var は不可）
 - 使用できるのは英字・数字・アンダースコアのみ（例：test_var! は不可）
 - 最初の文字は英字またはアンダースコアでなければならない（例：1_var　は不可）

変数を作るときは、＝を使って値を割り当てます。

### Manipulating variables

You can always change the value assigned to a variable by overriding the previous value.

In the code cell below, we change the value of my_var from 3 to 100.

訳）変数に割り当てられている値は、以前の値を上書きすることでいつでも変更できます。
下のコードセルではmy_varの値を3から100に変更しています。

```python
# Set the value of a new variable to 3
my_var = 3

# Print the value assigned to my_var
print(my_var)

# Change the value of the variable to 100
my_var = 100

# Print the new value assigned to my_var
print(my_var)
```

Note that in general, whenever you dedine a variable in a code cell, all of the code cells that follow also have access to the variable.
For instance, we use the next code cell to access the values of my_var (from the code cell above) and test_var (from earlier in this tutorial).

訳）一般的にコードセルで変数を定義すると、その後のすべてのコードセルからもその変数を参照できます。
例えば、次のコードセルでは、上のコードセルで定義したmy_varの値と、このチュートリアルの冒頭で定義したtest_varを参照しています。

```python
print(my_var)
print(test_var)
```

The next code cell tells Python to increase the current value of my_var by 3.
To do this, we still need to use my_var = like before.
And also just like before, the new value we want to assign to the variable is to the right of the = sign.

訳）次のコードセルは、my_varの現在の値に3を加えるようにPythonで指示します。
これを行うには、以前と同じようにmy_var = を使う必要があります。
また、以前と同じく、変数に割り当てたい新しい値は＝の右側に置きます。

```python
# Increase the value by 3
my_var = my_var + 3

# Print the value assigned to my_var
print(my_var)
```

### Using multiple variables

It's common for code to use multiple variables.
This is especially useful when we have to do a long calculation with multiple inputs.

In the next code cell, we calculate the number of seconds in four years.
This calcutation uses five inputs.

訳）コードで複数の変数を使うことはよくあります。
特に複数の入力を使った長い計算をおこなう場合に役立ちます。

次のコードセルでは、4年間の秒数を計算します。
この計算では、5つの入力を使用しています。

```python
# Create variables
num_years = 4
days_per_year = 365
hours_per_day = 24
mins_per_hour = 60
secs_per_min = 60

# Calculate number of seconds in four years
total_secs = secs_per_min * mins_per_hour * hours_per_day * days_per_year * num_years
print(total_secs)
```

As calculated above, there are 126144000 seconds in four years.

Note it is posssible to do this calculation without variables as just 60 * 60 *24 * 365 * 4, but it is much harder to check that the calculation without variables does not have some error, because it is not as readable.
When we use variables (such as num_years, days_per_year, etc), we can better keep track of each part of the calculation and more easily check for and correct any mistakes.

Note that it is particularly useful to use variables when the values of the inputs can change.
For instance, say we want to slightly improve our estimate by updating the value of the number of days in a year from 365 to 365.25, to account for leap years.
Then we can change the value assigned to days_per_year without changing any of the other variables and redo the calcultation.

訳）上で計算したように、4年間は126144000秒になります。

変数なしに60*60*24*365*4のように計算することも可能です。しかし変数を使わないと読みやすさが低く、誤りがないか確認するのが難しくなります。num_years,days_per_yearなどといった変数を使うときで、計算の各部分を追いやすくなり、ミスの修正や確認が簡単になります。

さらに、入力値が変わる可能性がある場合には、変数を使うことが特に有用です。
例えば、うるう年を考慮するために1年の日数を365日から365.25日へ更新して計算をより正確にしたいとします。
その場合、他の変数を変えることなくdays_per_yearの値だけを書き換えれば、計算をやり直すことができます。

```python
# Update to include leap years
days_per_year = 365.25

# Calculate number of seconds in four years
total_secs = secs_per_min * mins_per_hour * hours_per_day * days_per_year * num_years
print(total_secs)
```

**Note**: You might have noticed the .0 added at the end of the number, which might look unnecessary.
This is caused by the fact that in the second calculation, we used a number with a fractional part (365.25), whereas the first calculation multipled just numbers with no fractional part.
You'll learn more about this in Lesson 3, when we cover data types.

訳）注意：数字の最後に.0と付いていることに気づいたかもしれません。これは一見すると不要に見えるかもしれません。
この".0"は、2つ目の計算で少数（365.25）を使ったことが原因です。一方、最初の計算では整数だけを使っていました。
この違いについては、データ型を扱うレッスン３で詳しく学びます。

## Debugging

One common error when working with variables is to accidentally introduce typos.
For instace, if we spell hours_per_day as hours_per_dy, Python will error with message NameError: name 'hour_per_dy' is not defined.

訳）変数を扱う際によくあるエラーがうっかりタイピングミスしてしまうことです。
例えば、hours_per_dayをhours_per_dyと入力してしまうと、Pythonは"名前エラー：名前'hours_per_dyは定義されていない'"というエラーメッセージを出します。

```python
print(hours_per_dy)
```

When you see NameError like this, it's an indication that you should check how you have spelled the variable that it references as "not defined".
Then, to fix the error, you need only correct the spelling.

訳）このようなNameErrorを見たときは、"not defined"と表示されている変数の綴りを確認すべきだというサインです。
そして、エラーを直すには綴りを正しく修正するだけで済みます。

```python
print(hours_per_day)
```
