## Intro to functions: a simple example

We begin with a simple example of a fuction.
The add_three() function below accepts any number, adds three to it, and then returns the result.

訳）関数の簡単な例から始めます。
下のadd_three()関数は、どんな数字でも受け取り、それに3を足した値を返します。

```python
# Define the function
def add_three(input_var):
    output_var = input_var + 3
    return output_var
```

Every function is composed of two pieces: a header and body.

訳）どんな関数もヘッダーとボディの2つで構成しています。

```python
from IPython.display import Image
Image(r"function_img1.png")
```

### Header

The function **header** defines the name of the function and its argument(s).

- Every function header begins with def , which tells Python that we are about to define a function.
- In the example, the function name is add_three.
- In the example, the argument is input_var. The **argument** is the name of the variable that will be used as input to the function. It is always enclosed in parentheses that appear immediately after the name of the function. (Note that a function can also have no arguments, or it can have multiple arguments. You'll see some examples of this later in the lesson.)
- For every function, the parenthese enclosing the function argument(s) must be followed by a colon : .

訳）関数の**ヘッダー**では、関数名と引数を定義します。

- すべての関数ヘッダーはdefで始めます。これはPythonにこれから関数を定義することを知らせるためです。
- この例では関数名はadd_threeです。
- この例では引数はinput_varです。**引数**とはその関数に入力される値を受け取るための変数名です。関数名の直後に置かれた括弧の中に書きます。（引数がない関数もありますし、複数の引数を持つ関数もあります。これについてはレッスン後半で例がでてきます。）
- すべての関数で、引数を囲んむ括弧の後にはコロン:を引ける必要があります。

### Body

The function **body** specifies the work that the function does.

- Every line of code in the function body must be indented exactly four spaces. You can do this by pushing the space bar four times, or by hitting the "Tab" button once on your keyboard. (As you learn more about Python, you may need to indent your code by more than four spaces, but you'll learn more about that later in the course.)
- The function does its work by running all of the indented lines from top to bottom.
    - It takes the argument as input, which in the example is input_var.
    - The function creates a new variable output_var with the calculation output_var = input_var + 3.
    - Then, the final line of code, called the return statement, just returns the value in ouput_var as the function's output.

The code cell above just defines the function, but does not run it. The details of the function body will make more sense after the next code cell, when we actually run the function.

訳）関数のボディでは、関数で行う作業を指定します。

- 関数のボディ内のすべてのコード行は、正確に4つのスペース分インデントしなければなりません。キーボードのスペースバーを4回押すか、タブキーを1回押すことでできます。（Pythonをより学ぶと、4つスペース以上のインデントが必要になる場合がありますが、その詳細は後のコースで学びます。）
- 関数はインデントされたすべての行を上から下へ順に実行することで作業を行います。
    - 引数は入力として使われます。この例ではinput_varです。
    - 関数では、計算output_var = input_var + 3 を行う新しい変数out_varを作成します。
    - 最後の行であるreturn文は、output_varの値を関数の出力として返します。

上のコードセルは関数を定義しただけで、まだ実行はされていません。関数を実際に実行した後、ボディの詳細がどのように動作するかを理解できます。

## How to run (or "call") a funtion

When we run a function, it can also be referred to as "calling" the function.

In the code cell below, we run the function with 10 as the input value.
We define a new variable new_number which is set to the output of the function.

訳）関数を実行することは、関数を呼び出すとも言います。

下のコードセルでは、入力値として10を使って関数を実行します。
関数の出力は、新しい変数new_numberに代入されます。

```python
# Run the function with 10 as input
new_number = add_three(10)

# Check that the value is 13, as expected
print(new_number)
```

```python
Image(r"function_img2.png")
```

In more detail,

- add_three(10) is the value that we get as output when we supply 10 as the value for input_var and call the add_three() function. When the function runs, it runs all of the code in its body, from top to bottom:
    - It first calculates output_var = input_var + 3, which sets output_var = 13.
    - The final line of code is the return statement, which returns the value of output_var, which is 13.
- By setting new_number = add_three(10), we set new_number = 13.

**Note**: When we casually refer to the add_three() function in this tutorial, we use empty closing parentheses after the function name. 
This is consistent with how people generally write explanations of Python code, and the empty parentheses just make it clear that we are referring to a function, as opposed to a variable or another Python object.
These parentheses should always be empty, even if the function has arguments.

訳）より詳細に説明すると、

- add_three(10)はinput_varに10を渡してadd_three()関数を呼び出したときに得られる出力値です。関数が実行されると、関数本体コードが上から下へ順番に実行されます。
    - まずにoutput_var = input_var + 3　が計算され、output_var = 13　になります。
    - 最後の行にあるreturn文は、new_number = 13　を返します。
- new_number = add_three(10) と代入することで new_number = 13 に設定されます。

**注意**: このチュートリアルでは add_three() 関数に気軽に言及するとき、関数名の後に空の括弧をつけて書きます。
これは一般的なPythonの説明文でよく使われる慣習で、空の括弧をつけることで「これは変数ではなく関数である」ことを明確にするためです。
関数に引数がある場合でも、説明中では空の括弧のままにするのが普通です。

## A more complex example

Now that you understand the basics, we can move on to an example with a longer calculation.

Say you are helping a friend to calculate their weekly paycheck after taxes.

- They're in a 12 % tax bracket (in other words, 12% of their salary is taken for taxes, and they only take home 88%), and
- They're paid hourly, at a rate of $15/hour.

The function below calculates the paycheck based on the number of hours worked. The function is more complicated than with the first example, because the function has more lines of code and comments. Similar to the example above, the function has a single argument (num hours). In the function body, we:

- Use the value for num_hours to specify the value for a new variable pay_pretax.
- Use the value of pay_pretax to specify the value for a new variable pay_aftertax.
- Return the value of the pay_aftertax variable.

訳）基礎は理解したのでより長い計算例に進みます。

例えば、友人の税引き後の週給を計算する手伝いをしているとしましょう。

- 彼らは税率12%の階層に属しています。（言い換えれば、給料の12%が税金として差し引かれ、手取りは88%になります。）
- 彼らは時給制で雇われており、時給は$15です。

下の関数は、働いた時間数に基づいて給料(手取り)を計算します。この関数は最初の例より複雑です。なぜなら関数はコード行数とコメントが多いためです。上の例と同様に、この関数は1つの引数(num_hours)を受け取ります。関数本体では次の処理を行います。

- num_hoursの値を使って、新しい変数pay_pretaxの値を設定する。
- pay_pretaxの値を使って、新しい変数pay_aftertaxの値を設定する。
- pay_aftertaxの値を関数の戻り値として返す。

```python
def get_pay(num_hour):
    # Pre-tax pay, based on receiving $15/hour
    pay_pretax = num_hour * 15
    # After-tax pay, based on being in 12% tax bracket
    pay_aftertax = pay_pretax * (1 - .12)
    return pay_aftertax
```

We call this function the same way we called the first function.The next code cell calculates the paycheck, based on working 40 hours. (After taxes, it is $528.)

訳）この関数は、最初の関数と同様の方法で呼び出します。次のコードセルは40時間働いていた場合の給料(手取り)計算します。（税引き後の受取額は$528です。）

```python
# Calculate pay based on working 40 hours
pay_fulltime = get_pay(40)
print(pay_fulltime)
```

To quickly calculate pay based on a different number of hours worked, you need to supply the function with a different number. For instance, say your friend works 32 hours. (Then, they get $422.40.)

訳）異なる労働時間に基づく給料を素早く計算するには、関数に異なる数字を指定する必要があります。例えば友人が32時間働いたとした場合。（その場合は$422.40を受け取ります）

```python
pay_fulltime = get_pay(32)
print(pay_fulltime)
```

Because you wrote a function, you can calculate pay for different hours without having to rewrite all of the code in the calculations all over again.

Functions can help you to avoid errors in your code, and you save a lot of time. In general, when coding, you should aim to write as little as possible, because each time you type out a calculation, it's another opportunity to accidentally introduce a typo or error.

訳）関数を書いておけば、計算コードを毎回書き直さなくても、異なる労働時間に対する給料を計算できます。

関数を使うことでコードの誤りを避けやすくなり、作業時間を大幅に削減できます。一般的に、プログラミングではできるだけコードを書く量を減らすことが望ましいです。なぜなら、毎回計算式を入力するたびに、タイポやミスを誤っ起こす可能性が増えるためです。

## Variable "scope"

Variables defined inside the function body cannot be accessed outside of the function. For instance, the next code cell error, because pay_aftertax only exists inside the function.

訳）関数内部で定義された変数は関数外でアクセスできないです。例えば、次のコードセルはエラーになります。なぜならpay_aftertaxが関数内のみ存在するためです。

```python
print(pay_aftertax)
```

You will get the same error if you try to print pay_pretax or num_hours. For this reason, if you need any information from a function, you need to make sure that appears in the return statement at the end of the function.

We refer to a variable's **scope** as the part of the code where it is accessible. Variables defined inside a function (like pay_aftertax) have a local scope of that function only. However, as you've seen, variables defined outside all functions (like pay_parttime) have a global scope and can be accessed anywhere.

訳）pay_pretaxやnum_hoursを出力する場合も同じエラーが発生します。そのため、関数外で使いたい情報がある場合は、その情報が関数末尾にあるreturn文に含まれていることを確認する必要があります。
変数のスコープとは、その変数にアクセスできるコードの範囲のことを指します。関数内で定義された(pay_aftertaxのような)変数はその関数内のローカル変数です。しかし、関数外で定義された(pay_parttimeのような)変数はグローバル変数でどこでもアクセス可能です。

## Functions with multiple arguments

So far, you have learned how to define a function with just one argment. To define a function with multiple arguments, you only need to add more arguments inside the parentese in the function head and separate them with a comma.

We do this with the get_pay_with_more_inputs() function belowe, which calculates a weekly paycheck based on three arguments:

- num_hours - number of hours worked in one week
- hourly_wage - the hourly wage (in $/hour)
- tax_bracket - percentage of your salary that is removed for taxes

訳）これまで、引数1つを持つ関数をどのように定義するか学びました。複数の引数を定義するためには、関数ヘッダーの括弧内にコンマ区切りで複数の変数を加える必要があります。

３つの引数に基づいて週給が計算される下記のようなget_pay_with_more_inputs()関数を使用します。
- num_hours - 1週間の労働時間
- hourly_wage - 時給($/hour)
- tax_bracket - 給料から税金として差し引かれる割合

```python
def get_pay_with_more_inputs(num_hours, hourly_wage,tax_bracket):
    # Pre-tax pay
    pay_pretax = num_hours * hourly_wage
    # After-tax pay
    pay_aftertax = pay_pretax * (1 - tax_bracket)
    return pay_aftertax
```

Then, to call the function, you need to provide one value for each input, again separated by a comma.
In the code cell below, we calculate the pay after taxes for someone who works 40 hours, makes $24/hour, and is in a 22% tax bracket.

訳）関数を呼び出すために、各入力値に対する値をコンマで区切って渡す必要があります。
下のコードセルでは労働40時間、時給$24、税率22%で税引き後の給料を計算しています。

```python
higher_pay_aftertax = get_pay_with_more_inputs(40,24,.22)
print(higher_pay_aftertax)
```

The following code cell gives the same result as when we ran get_pay(40), because hourly_wage is set to 15, and tax_bracket is set to 12%, which lines up with how we designed get_pay.

訳）次のコードセルはget_pay(40)を実行したときと同じ結果を返します。なぜならhourly_wageを15、tax_bracketを12%と設定しており、これはget_payを設計したときの条件と一致しています。

```python
same_pay_fulltime = get_pay_with_more_inputs(40,15,.12)
print(same_pay_fulltime)
```

Depending on how we plan to use this new function get_pay_with_more_inputs(), it can be more useful than the original function get_pay(), because it addressed more cases. Instead of potentially incorrectly assuming the hourly wage and tax bracket, the new function allows the user to specify the correct values. But, if you're sure the hourly wage tax bracket won't need to change, the new function is just more complicated than necessay. In general, when defining functions, you'll need to consider how much flexibility you need, based on your use case.

訳）新しい関数のget_pay_with_more_inputs()をどのように使うかによって、元のget_pay()よりもっと役立ちます。なぜならより多くのケースに対応しているからです。hourly_wageとtax_bracketを誤って推定する可能性はあるますが、新しい関数は使用者が正しく値を指定することを許可しています。しかし時給や税率を変更する必要がない場合、新しい関数は必要以上に複雑になります。一般的に関数を定義する際、用途に応じてどのくらい柔軟性が必要か考える必要があります。
