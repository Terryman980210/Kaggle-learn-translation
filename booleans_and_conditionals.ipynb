## Booleans
Python has a type of variable called bool.
It has two possible values: True and False.

訳）Pythonにはboolと呼ばれる真偽値型があります。
この型が取り得る値はTrueとFalseの2つです。

```python
x = True
print(x)
print(type(x))
```

Rather than putting True or False directly in our code, we usually get **boolean values** from boolean operators.
There are operators that answer yes/no questions.
We'll go through some of these operators below.

訳）コードにTrueもしくはFalseを直接書く代わりに、通常は真偽演算子から真偽値を取得します。
それらの演算子はyes/noで答えられる条件を評価します。
これから、これらの演算子をいくつか見ていきます。

```python
def can_run_for_president(age):
    """Can someone of the given age run for president in the US?"""
    # The US Constitution says you must be at least 35 years old
    return  age >= 35

print("Can a 19-year-old run for president", can_run_for_president(19))
print("Can a 45-year-old run for president", can_run_for_president(45))
```

Comparisons frequently work like you'd hope

訳）比較は期待通りに動作します

```python
3.0 == 3
```

But sometimes they can be tricky

訳）しかし時々、トリッキーに動作することもあります。

```python
'3' == 3
```

Comparison operators can be combined with the arithmetic operators we've already seen to express a virtually limitless range of mathematical tests.
For example, we can check if a number is odd by checking that the modulus with 2 returns 1:

訳）比較演算子はこれまでに見たような算術演算子を組み合わせて、事実上無限の数学的テストを表現することができます。
例えば、数が奇数かどうかは、2で割った余りが1になるかどうかを確認することで調べられます。

```python
def is_odd(n):
    return (n % 2) == 1

print("Is 100 odd?", is_odd(100))
print("Is -1 odd?", is_odd(-1))
```

Remember to use == instead of = when making comparisons.
If you write n == 2 you are asking about the value of n.
When you write n = 2 you are changing the value of n.

訳）比較する際は=の代わりに==を使用することを忘れないでください。
n == 2と記述した場合、nの値が等しいかどうか確認しています。
一方、n = 2 と記述した場合は、nに値を代入しています。
