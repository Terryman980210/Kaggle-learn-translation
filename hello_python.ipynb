訳)Pythonは英国劇団の「モンティパイソン」から名付けられました。そこで、最初のPythonプログラムは彼らの「Spam」に関するコントへのオマージュを作ります。
ちょっとしたお遊びとして、まずは下のコードをざっと読んでみて、実行すると何が起きるか予想してください。

```python
spam_amount = 0
print(spam_amount)

# Ordering Spam, egg, Spam, bacon, and Spam (4 more servings of Spam)
spam_amount = spam_amount + 4

if spam_amount > 0:
    print("But I don't wnt ANY spam")

viking_song = "Spam " * spam_amount
print(viking_song)
```

There's a lot to unpack here!
This silly program demonstrates many important aspects of what Python code looks like and how it works.
Let's review the code from top to bottom.

訳）ここには、説明すべき点がたくさんあります。
このちょっとしたプログラムは、Pythonコードの書き方や、どのように動作するかといった重要なポイントを数多く示しています。
上から下まで順にコードを確認しましょう。

```python
spam_amount = 0
```

**Variable assignment:** Here we create a variable called spam_amount and assign it the value of 0 using = , which is called the assignment operator.

**Note**: if you've programmed in certain other languages (like Java or C++), you might be noticing some things Python doesn't require us to do here.
- we don't need to "declare" spam_amount berfore assigning to it
- we don't need to tell Python what type of value spam_amount is going to refer to. In fact, we can even go on to reassign spam_amount to refer to a different sort of thing like a string or a boolean.

訳)**変数の代入:** ここでは、spam_amountという変数を作成し、代入演算子=を使用して0の値を代入します。

**注釈**:　JavaやC++のような他の言語でプログラミングしたことがある場合と、Pythonではいくつかのことが不要になります。
- spam_amountを代入する前に、"宣言"する必要がありません。
- spam_amountの値がどのような型を参照するかを、あらかじめPythonに伝える必要もありません。実際にspam_amountに文字列や真偽値のような異なる型を第代入することもできます。

```python
print(spam_amount)
```

**Function calls:** print is a Python function that displays the value passed to it on the screen.
We call functions by putting parentheses after their name, and putting the inputs (or arguments) to the function in those parentheses.

訳）**関数の呼び出し:** printは画面に値を表示するPythonの関数です。
関数は、関数名の後に括弧を付け、その中に入力値(引数)を書くことで呼び出します。

```python
# Ordering Spam, egg, Spam, Spam, becon, and Spam (4 more servings of Spam)
spam_amount = spam_amount + 4
```

The first line above is a comment.
In Python, comments begin with the # symbol.

Next we see an example of reassignment.
Reassigning the value of an existing variable looks just the same as creating a variable it still uses the = assignment operator.

In the case, the value we're assgining to spam_amount involves some simple arithmetic on its previous value.
When it encounters this line, Python evaluates the expression on the right-hand-side of the = (0+4=4), and then assigns that value to the variable on the left-hand-side.

訳）上の1行目はコメントです。
Pythonでは、コメントは#記号で始まります。

次は再代入する例を見てみましょう。
すでに存在する変数に再代入する値は変数を作成したときと同じように代入演算子＝を使用します。

この場合、spam_amountに代入している値は前の値を使った単純な計算になっています。
この行が実行されると、Pythonは=の右辺に書いた(0+4=4)を評価し、その結果を左辺の変数に代入します。

```python
if spam_amount > 0:
    print("But I don't want ANY spam!")

viking_song = "Spam Spam Spam"
print(viking_song)
```
